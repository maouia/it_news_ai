# Automate the webscraping

```python
def get_data_from_page(url):
    page = requests.get(url)
    soup_1 = BeautifulSoup(page.text,'html.parser')
    
    
    body = soup_1.find_all('p')
    all_body=''
    for b in body :
        all_body+=re.sub(r'[^\w]', ' ', b.text)
    
    title = re.sub(r'[^\w]', ' ', soup_1.find('h1',{'class':'entry-title'}).text)

    image = soup_1.find('img',{'class':'entry-thumb'}).get('src')
    
  
    
    
    
    return [title,all_body,image]
    
    
    
    
    
```

```python
result  = get_data_from_page(href)
```

```python
result[1]
```

```python
titles = []
details = []
images = []
```

```python
for div in mydivs:
    a = div.find('a')
    href = a.get('href')
    all_data = get_data_from_page(href)
    titles.append(all_data[0])
    details.append(all_data[1])
    images.append(all_data[2])
    
```

# Save CSV

```python
import os
import pandas as pd
```

```python
df = pd.DataFrame({'titles': titles, 'details': details, 'images': images})


csv_file_path = "./output.csv"

# Append the DataFrame to the CSV file if it already exists, otherwise create it
if os.path.exists(csv_file_path):
    df.to_csv(csv_file_path, mode='a', header=False, index=False)
else:
    df.to_csv(csv_file_path, index=False)

print("Data from text files has been added to the CSV file successfully.")
```
